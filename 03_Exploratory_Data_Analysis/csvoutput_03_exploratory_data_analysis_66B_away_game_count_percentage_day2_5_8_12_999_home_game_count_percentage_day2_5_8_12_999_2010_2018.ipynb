```python
def get_game_count_percentage_away_home_both_day_window(df_, day_window_):
    
    len(day_window_) - 1
    
    df_ = df_.set_index('game_date')
    
    df1_ = df_.groupby(['team_id', 'season'])['away_home_']\
              .rolling(day_window_).sum().reset_index()\
              .rename(columns={'away_home_':'home_game_count_day' + day_window_[0:len(day_window_) - 1]})    
    
    df1_['home_game_count_day' \
         + day_window_[0:len(day_window_) - 1]] = \
    df1_['home_game_count_day' + day_window_[0:len(day_window_) - 1]]#.astype('int64')
    
    
    df_.away_home_ = abs(df_.away_home_ - 1)
    
    df0_ = df_.groupby(['team_id', 'season'])['away_home_']\
             .rolling(day_window_).sum().reset_index()\
             .rename(columns={'away_home_':'away_game_count_day' + day_window_[0:len(day_window_) - 1]})
    
    df0_['away_game_count_day' \
         + day_window_[0:len(day_window_) - 1]] = \
    df0_['away_game_count_day' + day_window_[0:len(day_window_) - 1]]#.astype('int64')
    
    
    df_.away_home_ = abs(df_.away_home_ - 1)
    
    df2_ = df_.groupby(['team_id', 'season'])['away_home_']\
              .rolling(day_window_).count().reset_index()\
              .rename(columns={'away_home_':'game_count_day' + day_window_[0:len(day_window_) - 1]})
    
    df2_['game_count_day' \
         + day_window_[0:len(day_window_) - 1]] = \
    df2_['game_count_day' + day_window_[0:len(day_window_) - 1]]#.astype('int64')
    
    
    column_names_common_list01_ = [k for k in df0_.columns if k in df1_.columns]
    
    df01_ = pd.merge(df0_, df1_, on=column_names_common_list01_)
    
    column_names_common_list012_ = [k for k in df01_.columns if k in df2_.columns]
    
    df012_ = pd.merge(df01_, df2_, on=column_names_common_list012_)
    
    column_names_common_list012_ = [k for k in df01_.columns if k in df2_.columns]

    df012_['away_game_percent_day' \
         + day_window_[0:len(day_window_) - 1]] = \
    df012_['away_game_count_day' + day_window_[0:len(day_window_) - 1]] /\
    df012_['game_count_day' + day_window_[0:len(day_window_) - 1]].values
    
    
    df012_['home_game_percent_day' \
         + day_window_[0:len(day_window_) - 1]] = \
    df012_['home_game_count_day' + day_window_[0:len(day_window_) - 1]] /\
    df012_['game_count_day' + day_window_[0:len(day_window_) - 1]].values
    
    df_ = pd.merge(df_, df012_, on=['team_id', 'season', 'game_date'])
    
    
    return df_
```
