```python
#https://en.wikipedia.org/wiki/List_of_cities_by_GDP
df_city_gdp = rcr('city_gdp.csv')
p(df_city_gdp)
```

```python
def index_containing_substring(the_list, substring):
    for i, s in enumerate(the_list):
        if substring in s:
              return i
    return -1
```
