```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python
df_5bsa0918 = pd.read_csv(os.path.join('..', 'processed_data', '5boxscoreadvancedv2_2009_2018.csv'))
print(df_5bsa0918.shape)
p(df_5bsa0918)
```

```python
df_5bsa0918.PLAYER_NAME = df_5bsa0918.PLAYER_NAME.fillna('')
df_5bsa0918.MIN = df_5bsa0918.MIN.fillna('00:00')
df_5bsa0918.MIN = pd.to_timedelta('00:' + df_5bsa0918.MIN).astype('timedelta64[s]').astype('float64') / 60
df_5bsa0918.TEAM_ID = df_5bsa0918.TEAM_ID.fillna(0).astype('int64')
print(df_5bsa0918.shape)
p(df_5bsa0918)
```

```python
path = os.path.join('..', 'processed_data', '3game_id_game_date_team_id.csv')
path
df_3gigdti = pd.read_csv(path, parse_dates=['game_date'])
df_3gigdti0918 = df_3gigdti[(df_3gigdti.season == '2009-10') |
                            (df_3gigdti.season == '2010-11') |
                            (df_3gigdti.season == '2011-12') |
                            (df_3gigdti.season == '2012-13') |
                            (df_3gigdti.season == '2013-14') |
                            (df_3gigdti.season == '2014-15') |
                            (df_3gigdti.season == '2015-16') |
                            (df_3gigdti.season == '2016-17') |
                            (df_3gigdti.season == '2017-18')]
print(df_3gigdti0918.shape)
p(df_3gigdti0918)
```
