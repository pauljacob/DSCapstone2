```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_):
    df_ = df_.sort_values(column_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of list
def pdc(dict_):
    print(len(dict_))
    return dict_
```

```python
df = rcp('51stat_cma_diff_2010_2018.csv', ['game_date'])
df = sr(df, 'game_date')
p(df)
```

```python
column_names_numeric_list = list(df.select_dtypes('number').columns)
pl(column_names_numeric_list)
```

```python
sm_column_names_number_list_cleaned = ['season', 'matchup'] + [k for k in column_names_number_list if
                                                                (not '_cma_' in k)
                                                               and (k not in ['game_id', 'season_year'])
                                                               and (not 'team_id' in k)]
pl(sm_column_names_number_list_cleaned)
```

```python
df_sm_numeric = df.loc[:,sm_column_names_number_list_cleaned]
p(df_sm_numeric)
```

```python
df_sm_cma = df_sm_numeric.groupby(['season', 'matchup']).apply(lambda x: x.rolling(12, 1).mean().shift())
df_sm_cma
```

```python
#fillna with 0 in preparation for taking the difference of season matchup cma's
df_sm_cma = df_sm_cma.fillna(0)
p(df_sm_cma)
```

```python
column_names_sm_cma_list = list(df_sm_cma.columns)
pl(column_names_sm_cma_list)
```

```python
column_names_number_a_list = [k for k in column_names_sm_cma_list if '_a' in k]
pl(column_names_number_a_list)
```

```python
column_names_number_b_list = [k for k in column_names_sm_cma_list if '_b' in k]
pl(column_names_number_b_list)
```

```python
column_names_number_list = [k.split('_a')[0] for k in column_names_number_a_list]
pl(column_names_number_list)
```

```python
column_names_number_sm_cma_a_list = [k + '_sm_cma_a' for k in column_names_number_list]
pl(column_names_number_sm_cma_a_list)
```

```python
column_names_number_sm_cma_b_list = [k + '_sm_cma_b' for k in column_names_number_list]
pl(column_names_number_sm_cma_b_list)
```

```python
column_names_number_sm_cma_a_dict = dict(zip(column_names_number_a_list, column_names_number_sm_cma_a_list))
pdc(column_names_number_sm_cma_a_dict)
```

```python
column_names_number_sm_cma_b_dict = dict(zip(column_names_number_b_list, column_names_number_sm_cma_b_list))
pdc(column_names_number_sm_cma_b_dict)
```

```python
#combine dictionaries
dict_temp = column_names_number_sm_cma_a_dict
dict_temp.update(column_names_number_sm_cma_b_dict)
column_names_number_sm_cma_ab_dict = dict_temp
pdc(column_names_number_sm_cma_ab_dict)
```

```python
df_sm_cma.rename(columns=column_names_number_sm_cma_ab_dict, inplace=True)
p(df_sm_cma)
```

```python
#matchup diff
df_sm_cma_diff = df_sm_cma[column_names_number_sm_cma_a_list] - df_sm_cma[column_names_number_sm_cma_b_list].values
p(df_sm_cma_diff)
```

```python
column_names_sm_cma_diff_list = [k + '_sm_cma_diff' for k in column_names_number_list]
pl(column_names_sm_cma_diff_list)
```

```python
column_names_sm_cma_diff_dict = dict(zip(column_names_number_sm_cma_a_list, column_names_sm_cma_diff_list))
pdc(column_names_sm_cma_diff_dict)
```

```python
df_sm_cma_diff = df_sm_cma_diff.rename(columns=column_names_sm_cma_diff_dict)
p(df_sm_cma_diff)
```

```python
df_sm_cma_diff = df_sm_cma_diff.reset_index()
df_sm_cma_diff = df_sm_cma_diff.rename(columns={'level_2': 'index'})
p(df_sm_cma_diff)
```

```python
df = df.reset_index()
p(df)
```

```python
df_merged = pd.merge(df, df_sm_cma_diff, on=['season', 'matchup', 'index'])
#df_merged.drop(columns=['index'], inplace=True)
p(df_merged)
```

```python
df_sm_cma_ = df_sm_cma.reset_index()
df_sm_cma_ = df_sm_cma_.rename(columns={'level_2': 'index'})
p(df_sm_cma_)
```

```python
df_merged2 = pd.merge(df_merged, df_sm_cma_, on=['season', 'matchup', 'index'])
df_merged2.drop(columns=['index'], inplace=True)
p(df_merged2)
```

```python
datapath = os.path.join('..', 'processed_data')

if not os.path.exists(datapath):
    os.mkdir(datapath)

datapath_df = os.path.join(datapath, '52stat_matchup_cma_diff_2010_2018.csv')
if not os.path.exists(datapath_df):
    df_merged2.to_csv(datapath_df, index=False)
```
