```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_):
    df_ = df_.sort_values(column_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of list
def pdc(dict_):
    print(len(dict_))
    return dict_
```

```python
df = rcp('52stat_matchup_cma_diff_2010_2018.csv', ['game_date'])
df = sr(df, 'game_date')
p(df)
```

```python
columns_names_all_list = list(df.columns)
pl(columns_names_all_list)
```

```python
column_names_number_list_pred = list(df.select_dtypes('number'))
pl(column_names_number_list_pred)
```

```python
column_names_cma_list_pred = [k for k in column_names_number_list_pred if '_cma' in k]
pl(column_names_cma_list_pred)
```

```python
columns_names_list_temp = [k for k in columns_names_all_list if (not k in column_names_cma_list_pred)]
columns_names_list_temp
```

```python
df_number_temp = df.loc[:, columns_names_list_temp].select_dtypes('number')
p(df_number_temp)
```

```python
column_names_df_number_temp_list = list(df_number_temp.columns)
pl(column_names_df_number_temp_list)
```

```python
column_names_remove_list = [k for k in column_names_df_number_temp_list if not k in ['game_id', 'team_id_a', 'team_id_b', 
                                                          'spread_a','spread_b','away_home_a', 'away_home_b',
                                                          'season_year']]
pl(column_names_remove_list)
```

```python
column_names_cleaning_list = [k for k in columns_names_all_list if not k in column_names_remove_list]
pl(column_names_cleaning_list)
```

```python
#not known prior to the start of the game
column_names_cleaning_list = [k for k in column_names_cleaning_list if
                              not k in ['duration_a', 'duration_b', 'MIN_a', 'MIN_b', 'wl_a',
                                        'wl_b', ]]
pl(column_names_cleaning_list)
```

```python
df_predictive = df.loc[:, column_names_cleaning_list]
p(df_predictive)
```

```python
datapath = os.path.join('..', 'processed_data')

if not os.path.exists(datapath):
    os.mkdir(datapath)

datapath_df = os.path.join(datapath, '53Aprediction_stats_2010_2018.csv')
if not os.path.exists(datapath_df):
    df_predictive.to_csv(datapath_df, index=False)
```
