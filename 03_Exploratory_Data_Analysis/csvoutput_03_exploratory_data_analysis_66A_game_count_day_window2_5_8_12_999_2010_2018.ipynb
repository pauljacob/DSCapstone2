```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_):
    df_ = df_.sort_values(column_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of list
def pdc(dict_):
    print(len(dict_))
    return dict_

#import my functions
import my_functions as f



def feature_engineer(df_, group_by_columns_, column_features_, operators_, window_size_, shift_num_=1):
    
    '''
    example operators: 'max', 'min', 'count', 'median', 'mean'
    '''
    df_.groupby([group_by_columns_])\
           [column_features_]\
           .rolling(window_size_,1).agg([operators_])\
           .shift(shift_num_)
```

```python
df = rcp('64_city_proper_metro_area_gdp_2010_2018.csv', ['game_date'])
p(df)
```

```python
df.loc[:, 'game_date_diff'] = df.groupby(['team_id', 'season']).game_date.diff(periods=1)
p(df)
```

```python
df.value_counts('game_date_diff')
```

```python
df = df.sort_values(['game_date', 'game_id'])
p(df)
```

```python
df.set_index(['game_date'], inplace=True)
p(df)
```

```python
df_game_count_day2 = df.groupby(['team_id', 'season'])['game_id'].rolling('2D')\
                                .count().reset_index().rename(columns={'game_id':'game_count_day_window2'})
p(df_game_count_day2)
```

```python
df_game_count_day5 = df.groupby(['team_id', 'season'])['game_id'].rolling('5D')\
                                .count().reset_index().rename(columns={'game_id':'game_count_day_window5'})
p(df_game_count_day5)
```

```python
#get number of games in the past 10 days
df_game_count_day8 = df.groupby(['team_id', 'season'])['game_id'].rolling('8D')\
                                .count().reset_index().rename(columns={'game_id':'game_count_day_window8'})
p(df_game_count_day8)
```

```python
df_game_count_day12 = df.groupby(['team_id', 'season'])['game_id'].rolling('12D')\
                                .count().reset_index().rename(columns={'game_id':'game_count_day_window12'})
p(df_game_count_day12)
```

```python
df_game_count_day999 = df.groupby(['team_id', 'season'])['game_id'].rolling('999D')\
                                .count().reset_index().rename(columns={'game_id':'game_count_day_window999'})
p(df_game_count_day999)
```

```python
column_names_common_list = [k for k in df_game_count_day999.columns if k in df_game_count_day12.columns]
pl(column_names_common_list)
```

```python
from functools import reduce
```

```python
data_frames = [df_game_count_day2, df_game_count_day5, df_game_count_day8, df_game_count_day12, df_game_count_day999]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=column_names_common_list,
                                            how='outer'), data_frames)
```
