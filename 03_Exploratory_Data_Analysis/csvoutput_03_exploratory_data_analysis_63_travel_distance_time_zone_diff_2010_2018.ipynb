```python
from pytz import timezone
import pandas as pd

def tz_diff(date, tz1, tz2):
    '''
    Returns the difference in hours between timezone1 and timezone2
    for a given date.
    '''
    date = pd.to_datetime(date)
    return (tz1.localize(date) - 
            tz2.localize(date).astimezone(tz1))\
            .seconds/3600 + (tz1.localize(date) - 
            tz2.localize(date).astimezone(tz1))\
            .days * 24
```

```python
team_city_time_zone = timezone('America/New_York')
game_team_city_time_zone = timezone('America/Chicago')

tz_diff('2017-06-01', team_city_time_zone, game_team_city_time_zone)

```

```python
time_zone_diff_list = [tz_diff('06-06-06', timezone(k1), timezone(k2)) for k1, k2 in zip(time_zone_list, game_time_zone_list)]
pl(time_zone_diff_list)
```
