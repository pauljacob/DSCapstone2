```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_):
    df_ = df_.sort_values(column_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of list
def pdc(dict_):
    print(len(dict_))
    return dict_

#import my functions
import my_functions as f
```

```python
df_away_home = rcp('21boxscoreadvancedv2_boxscores_team_away_home_2010_2018.csv', ['game_date'])
df_away_home = sr(df_away_home, 'game_date')
p(df_away_home)
```

```python
df_game_id_team_id = f.wide2tall(df_away_home, ['game_id', 'team_id'], ['_away', '_home'])
df_game_id_team_id = sr(df_game_id_team_id, 'game_date')
p(df_game_id_team_id)
```

```python
df_game_id_team_id['a'] = df_game_id_team_id.matchup.str.split(' vs. ').str[0].str.strip()

df_game_id_team_id['b'] = df_game_id_team_id.matchup.str.split(' vs. ').str[1].str.strip()


p(df_game_id_team_id)
```

```python
df_game_id_team_id.loc[df_game_id_team_id['a'] == df_game_id_team_id['TEAM_ABBREVIATION'], '_a'] = 1

df_game_id_team_id.loc[df_game_id_team_id['a'] != df_game_id_team_id['TEAM_ABBREVIATION'], '_a'] = 0


p(df_game_id_team_id)
```

```python
df_game_id_team_id.loc[df_game_id_team_id['b'] == df_game_id_team_id['TEAM_ABBREVIATION'], '_b'] = 1

df_game_id_team_id.loc[df_game_id_team_id['b'] != df_game_id_team_id['TEAM_ABBREVIATION'], '_b'] = 0


p(df_game_id_team_id)
```

```python
df_game_id_team_id = pd.concat([df_game_id_team_id, pd.get_dummies(df_game_id_team_id[['wl']])], axis=1)
p(df_game_id_team_id)
```

```python
df_game_id_team_id.rename(columns={'wl_L': 'loss', 'wl_W': 'win'}, inplace=True)
```

```python
p(df_game_id_team_id)
```

```python
df_loss_win = df_game_id_team_id.loc[:, ['game_id', 'team_id', 'season', 'loss', 'win']]
p(df_loss_win)
```

```python
df_losses_wins = df_loss_win.groupby(['team_id', 'season']).apply(lambda x: x.rolling(9999, 1).sum().shift())[['loss', 'win']]
p(df_losses_wins)
```

```python
df_losses_wins = df_losses_wins.reset_index()
p(df_losses_wins)
```

```python
df_losses_wins.rename(columns={'level_2': 'index', 'loss':'losses', 'win':'wins'}, inplace=True)
p(df_losses_wins)
```

```python
df_game_id_team_id = df_game_id_team_id.reset_index()
p(df_game_id_team_id)
```

```python
df_game_id_team_id_subset = df_game_id_team_id.loc[:, ['index', 'game_id', 'game_date', 'season', 'team_id', 'TEAM_ABBREVIATION', '_a']]
p(df_game_id_team_id_subset)
```

```python
df_losses_wins = pd.merge(df_game_id_team_id_subset, df_losses_wins, on=['index', 'season', 'team_id'])
p(df_losses_wins)
```

```python
df_losses_wins['win_pct'] = df_losses_wins.loc[:,'wins'] / \
                            (df_losses_wins.loc[:,'wins'] + df_losses_wins.loc[:,'losses'])
p(df_losses_wins)
```

```python
df_losses_wins.drop(columns='index', inplace=True)
p(df_losses_wins)
```

```python
df_losses_wins_ab = f.tall2wide2(df_losses_wins, ['game_id', 'game_date', 'season'], '_a')
p(df_losses_wins_ab)
```

```python
df_losses_wins_ab['opp_win_pct_a'] = df_losses_wins_ab['win_pct_b']
df_losses_wins_ab['opp_win_pct_b'] = df_losses_wins_ab['win_pct_a']

df_losses_wins_ab['opp_losses_a'] = df_losses_wins_ab['losses_b']
df_losses_wins_ab['opp_wins_a'] = df_losses_wins_ab['wins_b']

df_losses_wins_ab['opp_losses_b'] = df_losses_wins_ab['losses_a']
df_losses_wins_ab['opp_wins_b'] = df_losses_wins_ab['wins_a']

p(df_losses_wins_ab)
```

```python
df_losses_wins_opp_losses_wins = f.wide2tall(df_losses_wins_ab, ['game_id', 'team_id'], ['_a', '_b'])
p(df_losses_wins_opp_losses_wins)
```

```python
df_losses_wins_opp_losses_wins = sr(df_losses_wins_opp_losses_wins, ['game_date'])
p(df_losses_wins_opp_losses_wins)
```

```python
df_losses_wins_opp_losses_wins_rgd = df_losses_wins_opp_losses_wins.drop(columns=['game_date'])
p(df_losses_wins_opp_losses_wins_rgd)
```

```python
df_opp_losses_wins = df_losses_wins_opp_losses_wins_rgd.loc[:, ['team_id', 'season', 'opp_losses', 'opp_wins']]
p(df_opp_losses_wins)
```

```python
df_opp_losses_wins_csum = df_opp_losses_wins.groupby(['team_id', 'season']).apply(lambda x: x.rolling(9999, 1).sum())
p(df_opp_losses_wins_csum)
```

```python
df_opp_losses_wins_csum.drop(columns=['team_id'], inplace=True)
p(df_opp_losses_wins_csum)
```

```python
df_opp_losses_wins_csum = df_opp_losses_wins_csum.reset_index()
p(df_opp_losses_wins_csum)
```

```python
df_opp_losses_wins_csum.rename(columns={'level_2':'index', 
                                          'opp_losses':'opp_losses_csum',
                                          'opp_wins':'opp_wins_csum'}, inplace=True)
p(df_opp_losses_wins_csum)
```

```python
df_losses_wins_opp_losses_wins = df_losses_wins_opp_losses_wins.reset_index()
p(df_losses_wins_opp_losses_wins)
```

```python
df_lw_opplw_opplwcsum = pd.merge(df_losses_wins_opp_losses_wins,
                                          df_opp_losses_wins_csum,
                                          on=['team_id', 'season', 'index'])
p(df_lw_opplw_opplwcsum)
```

```python
#bit flip
df_lw_opplw_opplwcsum['a_b_'] = abs(df_lw_opplw_opplwcsum['a_b_'] - 1)
p(df_lw_opplw_opplwcsum)
```

```python
df_lw_opplw_opplwcsum.drop(columns=['index'], inplace=True)
```

```python
df_lw_opplw_opplwcsum_ab = f.tall2wide2(df_lw_opplw_opplwcsum,
                                  ['game_id', 'game_date', 'season'],
                                  'a_b_')
p(df_lw_opplw_opplwcsum_ab)
```

```python
df_lw_opplw_opplwcsum_ab.rename(columns={'opp_losses_csum_a':'opp_opps_losses_b',
                             'opp_wins_csum_a':'opp_opps_wins_b',
                             'opp_losses_csum_b':'opp_opps_losses_a',
                             'opp_wins_csum_b':'opp_opps_wins_a'}, inplace=True)
p(df_lw_opplw_opplwcsum_ab)
```

```python
p(df_lw_opplw_opplwcsum_ab)
```

```python
df_lw_opplw_opplwcsum_ab['opp_opps_win_pct_a'] = df_lw_opplw_opplwcsum_ab['opp_opps_wins_a'] \
                                                        / (df_lw_opplw_opplwcsum_ab['opp_opps_losses_a']
                                                         + df_lw_opplw_opplwcsum_ab['opp_opps_wins_a'])

df_lw_opplw_opplwcsum_ab['opp_opps_win_pct_b'] = df_lw_opplw_opplwcsum_ab['opp_opps_wins_b'] \
                                                        / (df_lw_opplw_opplwcsum_ab['opp_opps_losses_b']
                                                         + df_lw_opplw_opplwcsum_ab['opp_opps_wins_b'])
p(df_lw_opplw_opplwcsum_ab)
```

```python
df_lw_opplw_opplwcsum_ab['strength_of_schedule_a'] = (2 * df_lw_opplw_opplwcsum_ab['opp_win_pct_a']
                                                        + df_lw_opplw_opplwcsum_ab['opp_opps_win_pct_a']) \
                                                        / 3

df_lw_opplw_opplwcsum_ab['strength_of_schedule_b'] = (2 * df_lw_opplw_opplwcsum_ab['opp_win_pct_b'] 
                                                        + df_lw_opplw_opplwcsum_ab['opp_opps_win_pct_b']) \
                                                        / 3

p(df_lw_opplw_opplwcsum_ab)
```
