```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_):
    df_ = df_.sort_values(column_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of list
def pdc(dict_):
    print(len(dict_))
    return dict_
```

```python
df = rcp('21boxscoreadvancedv2_boxscores_team_away_home_2010_2018.csv', ['game_date'])
df = sr(df, 'game_date')
p(df)
```

```python
column_names_list = list(df.columns)
pl(column_names_list)
```

```python
column_names_away_list = [k for k in column_names_list if '_away' in k]
pl(column_names_away_list)
```

```python
column_names_home_list = [k for k in column_names_list if '_home' in k]
pl(column_names_home_list)
```

```python
column_names_away_home_list = column_names_away_list + column_names_home_list
column_names_game_level_list = [k for k in column_names_list if not k in column_names_away_home_list]
pl(column_names_game_level_list)
```

```python
52 + 52 + 6
```

```python
gi_column_names_away_list = ['game_id'] + column_names_away_list
pl(gi_column_names_away_list)
```

```python
gi_column_names_home_list = ['game_id'] + column_names_home_list
pl(gi_column_names_home_list)
```

```python
df_gia = df.loc[:,gi_column_names_away_list]
p(df_gia)
```

```python
df_gia.loc[:,'away_home'] = 0
p(df_gia)
```

```python
column_names_list_stripped = [k.split('_away')[0] for k in column_names_away_list]
pl(column_names_list_stripped)
```

```python
column_names_away_dict = dict(zip(column_names_away_list,column_names_list_stripped))
pdc(column_names_away_dict)
```

```python
column_names_home_dict = dict(zip(column_names_home_list,column_names_list_stripped))
pdc(column_names_home_dict)
```

```python
df_gia.rename(columns=column_names_away_dict, inplace=True)
p(df_gia)
```

```python
df_gih = df.loc[:,gi_column_names_home_list]
p(df_gih)
```

```python
df_gih.loc[:,'away_home'] = 1
p(df_gih)
```

```python
df_gih.rename(columns=column_names_home_dict, inplace=True)
p(df_gih)
```

```python
df_gi = pd.concat([df_gia, df_gih])
p(df_gi)
```

```python
df_gl = df.loc[:, column_names_game_level_list]
p(df_gl)
```

```python
df_giti = pd.merge(df_gi, df_gl, on='game_id', how='inner')
p(df_giti)
```

```python
df_giti = sr(df_giti, 'game_date')
p(df_giti)
```

```python
df_giti['a'] = df_giti.matchup.str.split(' vs. ').str[0].str.strip()
p(df_giti)
```

```python
df_giti['b'] = df_giti.matchup.str.split(' vs. ').str[1].str.strip()
p(df_giti)
```

```python
df_numeric = df_giti.select_dtypes('number')
df_numeric = df_numeric.drop(columns=['season_year'])
p(df_numeric)
```

```python
gi_column_names_numeric_list = list(df_numeric.columns)
pl(gi_column_names_numeric_list)
```

```python
column_names_numeric_list = [k for k in gi_column_names_numeric_list if not k in ['game_id']]
pl(column_names_numeric_list)
```

```python
column_names_numeric_season_list = column_names_numeric_list + ['season']
pl(column_names_numeric_season_list)
```

```python
df_temp = df_giti[column_names_numeric_season_list]
p(df_temp)
```

```python
df_temp.info()
```

```python
column_names_numeric_list_no_tis = [k for k in column_names_numeric_season_list if not k in ['team_id', 'season']]
pl(column_names_numeric_list_no_tis)
```

```python
df_sti_cma = df_temp.groupby(['season', 'team_id']).apply(lambda x: x.rolling(12, 1).mean().shift())[column_names_numeric_list_no_tis]
p(df_sti_cma)
```

```python
df_sti_cma = df_sti_cma.reset_index()
df_sti_cma.rename(columns={'level_2':'index'}, inplace=True)
p(df_sti_cma)
```

```python
column_names_list_cma = [k + '_cma' for k in column_names_numeric_list_no_tis]
pl(column_names_list_cma)
```

```python
column_names_cma_dict = dict(zip(column_names_numeric_list_no_tis, column_names_list_cma))
pdc(column_names_cma_dict)
```

```python
df_sti_cma.rename(columns=column_names_cma_dict,inplace=True)
p(df_sti_cma)
```

```python
df_giti = df_giti.reset_index()
p(df_giti)
```

```python
df_merged = pd.merge(df_giti, df_sti_cma, on=['index', 'team_id', 'season'], how='inner')
p(df_merged)
```

```python
#fillna cma columns with 0 (leading into taking the differences)

df_merged[column_names_list_cma] = df_merged[column_names_list_cma].fillna(0)
p(df_merged)
```

```python
#drop columns... 'index'
df_merged.drop(columns=['index'], inplace=True)
p(df_merged)
```
