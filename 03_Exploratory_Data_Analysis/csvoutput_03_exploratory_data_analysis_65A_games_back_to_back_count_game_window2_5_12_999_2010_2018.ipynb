```python
def get_group_by_sum(df_, group_by_columns_list_, window_size_max_min_):
    
    df_.loc[:, 'game_date_diff1'] = 0
    df_.loc[df.game_date_diff == pd.Timedelta(seconds=60*60*24), 'game_date_diff1'] = 1
    df_ = df_.reset_index()
    
    df_games_back_to_back_window10 = df.groupby(group_by_columns_list_)\
                                            ['game_date_diff1'].rolling(window_size_max_min_[0],
                                                                        window_size_max_min_[1]).sum().reset_index()
    
    
    gb_len = len(group_by_columns_list_)
    
    level_num = 'level_' + str(gb_len)
    
    games_back_to_back_window_num = 'games_back_to_back_count_game_window' + str(window_size_max_min_[0])
    
    df_games_back_to_back_window10 =  df_games_back_to_back_window10\
                                                .rename(columns = \
                                                {level_num:'index',
                                                 'game_date_diff1': games_back_to_back_window_num})
    
    merge_on_ = group_by_columns_list_ + ['index']
    
    df_ = pd.merge(df_, df_games_back_to_back_window10, on=merge_on_)
    
    return df_
    
    
```
