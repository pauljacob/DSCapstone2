```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python
#difference of those ending with _away or _home, but not including team_id? *stat* + _ah_diff

list_giah = [i for i in col_names_numeric if i not in ['team_id_away', 'team_id_home', 'season_year']]
list_giah
```

```python
len(list_giah)
```

```python
#get prefix to _away and _home and store as unique list

list_stats_h =  [i for i in list_giah if '_home' in i]
list_stats_h
```

```python
list_stats = [i.split('_home')[0] for i in list_stats_h]
list_stats
```

```python
len(list_stats)
```

```python
# _ah_diff
p(df)
```

```python
for stat in list_stats:
    df[stat + '_ah_diff'] = df[stat + '_away'] - df[stat + '_home']
```

```python
list_stat_ah_diff = []
for stat in list_stats:
    list_stat_ah_diff.append(stat + '_ah_diff') 
```

```python
#sort by date and reset index
df.sort_values('game_date', inplace=True)
df = df.reset_index(drop=True)

#groupby matchup and season and take cma of stat differences
df_stats_ah_diff_cma = df.groupby(['season', 'matchup'])[list_stat_ah_diff].expanding(min_periods=0).mean().reset_index()

#keep the original index
df_stats_ah_diff_cma.rename(columns={'level_2':'index'}, inplace=True)
p(df_stats_ah_diff_cma)
```
