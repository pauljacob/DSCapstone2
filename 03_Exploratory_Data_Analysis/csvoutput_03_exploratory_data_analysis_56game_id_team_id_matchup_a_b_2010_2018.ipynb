```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('display.max_colwidth', None)
pd.options.display.max_info_columns = 999

#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_):
    df_ = df_.sort_values(column_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of list
def pdc(dict_):
    print(len(dict_))
    return dict_

#import my functions
import my_functions as f
```

```python
df = rcp('21boxscoreadvancedv2_boxscores_team_away_home_2010_2018.csv', ['game_date'])
df = sr(df, 'game_date')
p(df)
```

```python
df_giti = f.wide2tall(df, ['game_id', 'team_id'], ['_away', '_home'])
p(df_giti)
```

```python
df_giti = sr(df_giti, 'game_date')
p(df_giti)
```

```python
df_giti['a'] = df_giti.matchup.str.split(' vs. ').str[0].str.strip()
p(df_giti)
```

```python
df_giti['b'] = df_giti.matchup.str.split(' vs. ').str[1].str.strip()
p(df_giti)
```

```python
df_giti.loc[df_giti['a'] == df_giti['TEAM_ABBREVIATION'], '_a'] = 1
p(df_giti)
```

```python
df_giti.loc[df_giti['a'] != df_giti['TEAM_ABBREVIATION'], '_a'] = 0
p(df_giti)
```

```python
df_giti.loc[df_giti['b'] == df_giti['TEAM_ABBREVIATION'], '_b'] = 1
p(df_giti)
```

```python
df_giti.loc[df_giti['b'] != df_giti['TEAM_ABBREVIATION'], '_b'] = 0
p(df_giti)
```

```python
column_names_list = list(df_giti.columns)
pl(column_names_list)
```

```python
column_names_list_matchup = [k for k in column_names_list if not k in column_names_list_dropped]
pl(column_names_list_matchup)
```

```python
df_giti_matchup = df_giti.loc[:,column_names_list_matchup]
p(df_giti_matchup)
```
