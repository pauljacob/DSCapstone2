## 4.4 Functions<a id='4.4_Imports'></a>

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_, ascending_=True):
    df_ = df_.sort_values(column_, ascending=ascending_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of dictionary
def pdc(dict_):
    print(len(dict_))
    return dict_
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='spread_a'), 
                                                    df.spread_a, test_size=0.3, 
                                                    random_state=47)
```

```python
scaler = StandardScaler()

```

```python
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
lm = LinearRegression().fit(X_train_scaled, y_train)
```

```python
y_train_pred = lm.predict(X_train_scaled)
y_test_pred = lm.predict(X_test_scaled)
```

```python

# r^2 - train, test
linear_regression_r2_score = r2_score(y_train, y_train_pred), r2_score(y_test, y_test_pred)
linear_regression_r2_score
```

```python
mean_absolute_error(y_train, y_train_pred), mean_absolute_error(y_test, y_test_pred)
```

```python
RF_pipe = make_pipeline(
    StandardScaler(),
    RandomForestRegressor(random_state=47)
)
```

```python
rf_default_cv_results = cross_validate(RF_pipe, X_train, y_train, cv=5)
```

```python
rf_cv_scores = rf_default_cv_results['test_score']
rf_cv_scores
```

```python
np.mean(rf_cv_scores), np.std(rf_cv_scores)
```

```python
n_est = [int(n) for n in np.logspace(start=1, stop=3, num=20)]
grid_params = {
        'randomforestregressor__n_estimators': n_est,
        'standardscaler': [StandardScaler(), None],
}
grid_params
```

```python
rf_grid_cv = GridSearchCV(RF_pipe, param_grid=grid_params, cv=5, n_jobs=-1)
```

```python
rf_grid_cv.fit(X_train, y_train)
```

```python
rf_grid_cv.best_params_
```

```python
rf_best_cv_results = cross_validate(rf_grid_cv.best_estimator_, X_train, y_train, cv=5)
rf_best_scores = rf_best_cv_results['test_score']
rf_best_scores
```

```python
np.mean(rf_best_scores), np.std(rf_best_scores)
```

```python
plt.subplots(figsize=(10, 5))
imps = rf_grid_cv.best_estimator_.named_steps.randomforestregressor.feature_importances_
rf_feat_imps = pd.Series(imps, index=X_train.columns).sort_values(ascending=False)
rf_feat_imps.plot(kind='bar')
plt.xlabel('features')
plt.ylabel('importance')
plt.title('Best random forest regressor feature importances');
```
