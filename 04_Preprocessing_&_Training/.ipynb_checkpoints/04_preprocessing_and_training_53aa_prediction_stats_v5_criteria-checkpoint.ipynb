## 4.4 Functions<a id='4.4_Imports'></a>

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None, ic_=None):
    if (pd_ == None) and (ic_ == None):
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    elif (pd_ != None) and (ic_ == None):
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    elif (pd_ != None) and (ic_ != None):
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_, index_col = ic_)
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), index_col = ic_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
    
#sort dataframe by column
def s(df_, column_):
    return df_.sort_values(column_)

#reset index and sort dataframe by column
def sr(df_, column_, ascending_=True):
    df_ = df_.sort_values(column_, ascending=ascending_)
    return df_.reset_index(drop=True)

#print length of list
def pl(list_):
    print(len(list_))
    return list_

#print length of dictionary
def pdc(dict_):
    print(len(dict_))
    return dict_
```

## 4.6 Get Indicator Variables<a id='4.6 Get_Indicator_Variables'></a>

```python
# Create dummy or indicator features for categorical variables Hint: you’ll need
# to think about your old favorite pandas functions here like get_dummies().

df.select_dtypes('object')
```

```python
column_names_categorical_list = list(df.select_dtypes('object').columns) + ['team_id_a', 'team_id_b', 'away_home_a']
pl(column_names_categorical_list)
```

```python
column_names_categorical_drop_list = [k for k in column_names_categorical_list if
                                     not k in ['away_home_a']]
pl(column_names_categorical_drop_list)
```

```python
df_indicator_variables = pd.get_dummies(df[column_names_categorical_list_cleaned])
p(df_indicator_variables)
```

```python
#cat drop
df = df.drop(columns=column_names_categorical_list)
p(df)
```

```python
df_numeric = df.select_dtypes('number')
p(df_numeric)
```

```python
column_names_numeric_list  = list(df_numeric.columns)
pl(column_names_numeric_list)
```

```python
column_names_numeric_remove_list = ['USG_PCT_sm_cma_diff', 'E_PACE_sm_cma_diff', 'PACE_sm_cma_diff', 'PACE_PER40_sm_cma_diff', 'duration_minutes_sm_cma_diff', 'min_sm_cma_diff']
pl(column_names_numeric_remove_list)
```
