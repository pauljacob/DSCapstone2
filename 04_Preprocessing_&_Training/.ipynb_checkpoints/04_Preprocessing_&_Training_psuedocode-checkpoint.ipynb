```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python
#########################################################################
# team, score, and betting data to be read in: df_team_score_betting

```

```python
df_game_betting_totals = rcp('4game_id_team_stats_betting_totals_2010_2018.csv')
p(df_game_betting_totals)
```

```python
df_lc_mean = rcp('16lc_mean_advanced_stat_2010_2018.csv')
p(df_lc_mean)
```

```python
df_lc_sum = rcp('16lc_sum_advanced_stat_2010_2018.csv')
p(df_lc_sum)
```

```python
df_lc_max = rcp('16lc_max_advanced_stat_2010_2018.csv')
p(df_lc_max)
```

```python
'''######################################
#df_lc_mean and df_lc_sum to be combined: df_lc_mean_sum

df_lc_mean_sum = pd.merge(df_lc_sum, df_lc_mean, on=['game_id', 'team_id', 'game_date', 'season'], how='inner')
df_lc_mean_sum'''
```

```python
from functools import reduce
data_frames = [df_lc_sum, df_lc_mean, df_lc_max]
```

```python
df_lc_sum_mean_max = reduce(lambda  left,right: pd.merge(left,right,on=['game_id', 'team_id', 'game_date', 'season'],
                                                how='inner'), data_frames)
p(df_lc_sum_mean_max)
```

```python
######################################
#df_lc_mean and df_lc_sum to be combined: df_lc_mean_sum

#merge with game_id's, team_id's for 2010-11 thru 2017-18 seasons by reading in 3game_id (game_id, team_id); merge left on 3game_id

#fill games with NaN for lost contribution with 0

```

```python
df_gitis = rcp('3game_id_game_date_team_id.csv')
p(df_gitis)
```

```python
df_gitis1018 = df_gitis[(df_gitis.season == '2010-11') | 
         (df_gitis.season == '2011-12') | 
         (df_gitis.season == '2012-13') | 
         (df_gitis.season == '2013-14') | 
         (df_gitis.season == '2014-15') | 
         (df_gitis.season == '2015-16') | 
         (df_gitis.season == '2016-17') | 
         (df_gitis.season == '2017-18')].reset_index(drop=True)
p(df_gitis1018)
```

```python
df_gitis1018 = df_gitis1018[['game_id', 'game_date', 'team_id', 'season']]
p(df_gitis1018)
```

```python
#get all game_id's, and team_id's for the 2010-11 thru 2017-18 seasons
df_lc_sum_mean_max_merged1 = pd.merge(df_gitis1018, df_lc_sum_mean_max, on=['game_id', 'game_date', 'team_id', 'season'], how='left')
df_lc_sum_mean_max_merged1
```

```python
df_lc_sum_mean_max_merged1.fillna(0, inplace=True)
```

```python
p(df_lc_sum_mean_max_merged1)
```

```python
#df_home
pd.merge(df_game_betting_totals, df_lc_sum_mean_max_merged1.rename(columns={'team_id': 'team_id_home'}), how='inner', on=['game_id', 'game_date', 'team_id_home', 'season'])
```

```python
col_list = list(df_lc_sum_mean_max_merged1.columns)
col_list
```

```python
col_list_ = [k for k in col_list if k not in ['game_id', 'game_date', 'season']]
col_list_home = [k + '_home' for k in col_list if k not in ['game_id', 'game_date', 'season']]
col_list_away = [k + '_away' for k in col_list if k not in ['game_id', 'game_date', 'season']]
```

```python
dh = dict(zip(col_list_, col_list_home))
da = dict(zip(col_list_, col_list_away))
dh
```

```python
df_game_betting_totals_lc_home = pd.merge(df_game_betting_totals, df_lc_sum_mean_max_merged1.rename(columns=dh), how='inner', on=['game_id', 'game_date', 'team_id_home', 'season'])
p(df_game_betting_totals_lc_home)
```

```python
df_game_betting_totals_lc_home_away = pd.merge(df_game_betting_totals, df_lc_sum_mean_max_merged1.rename(columns=da), how='inner', on=['game_id', 'game_date', 'team_id_away', 'season'])
p(df_game_betting_totals_lc_home_away)
```

```python
df_game_betting_totals_lc_home_away = df_game_betting_totals_lc_home_away.sort_values('game_date').reset_index(drop=True)
p(df_game_betting_totals_lc_home_away)
```

```python
pd.options.display.max_info_columns = 999
df_game_betting_totals_lc_home_away.info()
```

```python
df_tist = rcp('10team_id_season_team_2010_2018.csv')
p(df_tist)
```

```python
df_th = pd.merge(df_game_betting_totals_lc_home_away, df_tist.rename(columns={'TEAM_ID': 'team_id_home', 'Team':'Team_home'}), on=['team_id_home', 'season'], how='outer', indicator=True)
p(df_th)
```

```python
df_th.value_counts('_merge')
```

```python
df_thta = pd.merge(df_th.drop(columns=['_merge']), df_tist.rename(columns={'TEAM_ID': 'team_id_away', 'Team':'Team_away'}), on=['team_id_away', 'season'], how='outer', indicator=True)
p(df_thta)
```

```python
df_thta.value_counts('_merge')
```

```python
df_thta.drop(columns=['_merge'], inplace=True)
```

```python
p(df_thta)
```

```python
datapath = os.path.join('..', 'processed_data')

if not os.path.exists(datapath):
    os.mkdir(datapath)

datapath_df = os.path.join(datapath, '50betting_totals_lost_contribution_home_away_2010_2018.csv')
if not os.path.exists(datapath_df):
    df_thta.to_csv(datapath_df, index=False)
```
