```python

import pandas as pd
from matplotlib import pyplot as plt
#plt is called stateful because it has a current state in terms of what figure we are working with,
#what axes we are currently working with, things like that. We haven't talked much about figures and axes
#because we have been working with a single figure and a single axes object

#plt = Plot()

#plt.gcf() # used to get the current figure for a plot
#plt.gca() # used to get the current axes for a plot

#you can switch between different ones, that's the stateful way of doing it

#many people perfer to use the more object oriented approach when working with multiple figures and axes
#to do this, we can use the subplots method

plt.style.use('seaborn')

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']



figure, axes = plt.subplots(nrows=2, ncols=1)
#subplots create a figure and then specify a certain number of rows and columns of axes

print(axes)


axes.plot(ages, py_salaries, label='Python')
axes.plot(ages, js_salaries, label='JavaScript')

axes.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

axes.legend()

axes.set_title('Median Salary (USD) by Age')
axes.set_xlabel('Ages')
axes.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python

import pandas as pd
from matplotlib import pyplot as plt
#plt is called stateful because it has a current state in terms of what figure we are working with,
#what axes we are currently working with, things like that. We haven't talked much about figures and axes
#because we have been working with a single figure and a single axes object

def median_salary_by_age():
    plt.style.use('seaborn')

    data = pd.read_csv('data.csv')
    ages = data['Age']
    dev_salaries = data['All_Devs']
    py_salaries = data['Python']
    js_salaries = data['JavaScript']

    plt.plot(ages, py_salaries, label='Python')
    plt.plot(ages, js_salaries, label='JavaScript')

    plt.plot(ages, dev_salaries, color='#444444',
             linestyle='--', label='All Devs')

    plt.legend()

    plt.title('Median Salary (USD) by Age')
    plt.xlabel('Ages')
    plt.ylabel('Median Salary (USD)')

    plt.tight_layout()
    
    return plt

plt = median_salary_by_age()

#plt.show()
```
