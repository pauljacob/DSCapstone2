```python
def first_four_moves():
    one2eight_ = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
    onetwo_ = [k + 1 for k in range(2)]
    df_ = pd.DataFrame([[random.choice(one2eight), random.choice(onetwo)] for k in range(4)], index=[1, 2, 3, 4])
    df_.index.name = 'move#'
    df_.rename(columns={0:'file', 1:'rank'}, inplace=True)
    return df_
```

```python
def first_four_moves2():
    one2eight_ = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
    onetwo_ = [k + 1 for k in range(2)]
    df_ = pd.DataFrame([str(random.choice(one2eight)) + str(random.choice(onetwo)) for k in range(4)], index=[1, 2, 3, 4])
    df_.index.name = 'move#'
    df_.rename(columns={0:'square'}, inplace=True)
    return df_
```

```python
def first_four_moves3():
    one2eight_ = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
    onetwo_ = [k + 1 for k in range(2)]
    df_ = pd.DataFrame([str(random.choice(one2eight)) + str(random.choice(onetwo)) for k in range(4)], index=[1, 2, 3, 4])
    df_.index.name = 'move#'
    df_.rename(columns={0:'square'}, inplace=True)
    df_.reset_index(inplace=True)
    return df_
```
