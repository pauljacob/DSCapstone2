```python
def read_from_db_t2(text):
    c.execute(text)
    data = c.fetchall()
    print(data)
    for row in data:
        print(row)
```

```python
def read_from_db_t3(text):
    c.execute(text)
    data = c.fetchall()
    print(data)
    for row in data:
        print(row)
```

```python
def read_from_db_t4(text, table_name1, table_name2, table_name3):
    c.execute(text)
    data = c.fetchall()
    print(data)
    for row in data:
        print(row)
    c.execute('PRAGMA table_info(' + table_name1 + ');')
    data = c.fetchall()
    print(data)
    c.execute('PRAGMA table_info(' + table_name2 + ');')
    data = c.fetchall()
    print(data)
    c.execute('PRAGMA table_info(' + table_name3 + ');')
    data = c.fetchall()
    print(data)
```

```python
read_from_db_t3('SELECT * FROM Facilities')
```

```python
query1 = """SELECT f.name as facilityname,
	(m.firstname || ' ' || m.surname) as membername,
	(CASE WHEN m.memid != 0 THEN f.membercost
	ELSE f.guestcost END) AS cost
FROM `Bookings` as b
LEFT JOIN `Members` as m
ON m.memid = b.memid
LEFT JOIN `Facilities` as f
ON b.facid = f.facid
WHERE DATE(starttime) = '2012-09-14'
AND (m.memid = 0
AND guestcost * slots > 30)
OR (m.memid != 0 AND membercost * slots > 30)
ORDER BY cost DESC"""
```

```python
# A10: cost per customer per facility and sum
```

```python
query2 = """SELECT f.name as facilityname,
	(m.firstname || ' ' || m.surname) as membername,
	(CASE WHEN m.memid != 0 THEN f.membercost
	ELSE f.guestcost END) AS cost
FROM `Bookings` as b
LEFT JOIN `Members` as m
ON m.memid = b.memid
LEFT JOIN `Facilities` as f
ON b.facid = f.facid
WHERE DATE(starttime) = '2012-09-14'
AND (m.memid = 0
AND guestcost * slots > 30)
OR (m.memid != 0 AND membercost * slots > 30)
ORDER BY cost DESC"""
```

```python
read_from_db_t4(query2, 'Bookings', 'Facilities', 'Members')
```

```python
query3 = """SELECT f.name as facilityname,
	(m.firstname || ' ' || m.surname) as membername,
	(CASE WHEN m.memid != 0 THEN f.membercost * slots
	ELSE f.guestcost * slots END) AS cost,
    b.starttime,
    b.slots
FROM `Bookings` as b
LEFT JOIN `Members` as m
ON m.memid = b.memid
LEFT JOIN `Facilities` as f
ON b.facid = f.facid
ORDER BY cost DESC"""
```

```python
read_from_db_t4(query3, 'Bookings', 'Facilities', 'Members')
```

```python
query4 = """SELECT f.name as facilityname,
	SUM((CASE WHEN m.memid != 0 THEN f.membercost * slots
	ELSE f.guestcost * slots END)) AS totalrevenue
FROM `Bookings` as b
LEFT JOIN `Members` as m
ON m.memid = b.memid
LEFT JOIN `Facilities` as f
ON b.facid = f.facid
GROUP BY facilityname
ORDER BY totalrevenue DESC"""
```

```python
read_from_db_t4(query4, 'Bookings', 'Facilities', 'Members')
```

```python
query5 = """SELECT f.name as facilityname,
	CASE WHEN (SUM((CASE WHEN m.memid != 0 THEN f.membercost * slots
	ELSE f.guestcost * slots END)) < 1000) THEN SUM((CASE WHEN m.memid != 0 THEN f.membercost * slots
	ELSE f.guestcost * slots END)) ELSE 333 END AS totalrevenue
FROM `Bookings` as b
LEFT JOIN `Members` as m
ON m.memid = b.memid
LEFT JOIN `Facilities` as f
ON b.facid = f.facid
GROUP BY facilityname
ORDER BY totalrevenue DESC"""
```

```python
read_from_db_t4(query5, 'Bookings', 'Facilities', 'Members')
```

```python
query6 = """
SELECT *
FROM (SELECT f.name as facilityname,
    SUM((CASE WHEN m.memid != 0 THEN f.membercost * slots
	ELSE f.guestcost * slots END)) AS totalrevenue
FROM `Bookings` as b
LEFT JOIN `Members` as m
ON m.memid = b.memid
LEFT JOIN `Facilities` as f
ON b.facid = f.facid
GROUP BY facilityname
ORDER BY totalrevenue DESC) as subquery
WHERE subquery.totalrevenue < 1000"""
```

```python
read_from_db_t4(query6, 'Bookings', 'Facilities', 'Members')
```

```python
# Q12: Find the facilities with their usage by member, but not guests
query15 = """
SELECT m.memid,
    m.firstname || ' ' || m.surname as membername,
    SUM(b.slots)
FROM Members as m
LEFT JOIN Bookings as b
ON m.memid = b.memid
LEFT JOIN Facilities as f
ON f.facid = b.facid
WHERE m.memid != 0
GROUP BY m.memid"""
#m.recommendedby
#m.memid
```

```python
read_from_db_t4(query15, 'Bookings', 'Facilities', 'Members')
```

```python
# Q13: Find the facilities usage by month, but not guests 
query16 = """
SELECT strftime('%m', b.starttime) as month,
    SUM(b.slots) as totalslots
FROM Members as m
LEFT JOIN Bookings as b
ON m.memid = b.memid
LEFT JOIN Facilities as f
ON f.facid = b.facid
WHERE m.memid != 0
GROUP BY month"""
#m.recommendedby
#m.memid
```

```python
# Q13: Find the facilities usage by month, but not guests 
query19 = """
SELECT strftime('%m', bb.starttime) as month,
    ff.name,
    SUM(bb.slots) as totalslots
FROM (SELECT m.memid FROM `Members` as m WHERE m.memid != 0) as mm
LEFT JOIN (SELECT b.facid, b.memid, b.starttime, b.slots FROM `Bookings` as b) as bb
ON mm.memid = bb.memid
LEFT JOIN (SELECT f.facid, f.name FROM `Facilities` as f) as ff
ON ff.facid = bb.facid
GROUP BY ff.name"""
#m.recommendedby
#m.memid
```
