```python
import pandas as pd
import os
```

```python
df_nf = pd.read_csv(os.path.join('..', 'raw_data', 'nba_franchises.csv'))
df_nf.END = df_nf.END.fillna(9999)
df_nf = df_nf[(df_nf['FRANCHISE NAME'] != 'St. Louis Hawks') & 
              (df_nf['FRANCHISE NAME'] != 'Milwaukee Hawks') & 
              (df_nf['FRANCHISE NAME'] != 'Tri-Cities Blackhawks') &
              (df_nf['FRANCHISE NAME'] != 'New York Nets') &
              (df_nf['FRANCHISE NAME'] != 'Ft. Wayne Zollner Pistons') &
              (df_nf['FRANCHISE NAME'] != 'San Francisco Warriors') &
              (df_nf['FRANCHISE NAME'] != 'Philadelphia Warriors') &
              (df_nf['FRANCHISE NAME'] != 'San Diego Rockets') &
              (df_nf['FRANCHISE NAME'] != 'San Diego Clippers') &
              (df_nf['FRANCHISE NAME'] != 'Buffalo Braves') &
              (df_nf['FRANCHISE NAME'] != 'Minneapolis Lakers') &
              (df_nf['FRANCHISE NAME'] != 'Vancouver Grizzlies') &
              (df_nf['FRANCHISE NAME'] != 'New Orleans/Oklahoma City Hornets') &
              (df_nf['FRANCHISE NAME'] != 'Seattle SuperSonics') &
              (df_nf['FRANCHISE NAME'] != 'Syracuse Nationals') &
              (df_nf['FRANCHISE NAME'] != 'Kansas City Kings') &
              (df_nf['FRANCHISE NAME'] != 'Kansas City-Omaha Kings') &
              (df_nf['FRANCHISE NAME'] != 'Cincinnati Royals') &
              (df_nf['FRANCHISE NAME'] != 'Rochester Royals') &
              (df_nf['FRANCHISE NAME'] != 'New Orleans Jazz') &
              (df_nf['FRANCHISE NAME'] != 'Washington Bullets') &
              (df_nf['FRANCHISE NAME'] != 'Capital Bullets') &
              (df_nf['FRANCHISE NAME'] != 'Baltimore Bullets') &
              (df_nf['FRANCHISE NAME'] != 'Chicago Zephyrs') &
              (df_nf['FRANCHISE NAME'] != 'Chicago Packers')]

df_nf.reset_index(drop=True, inplace=True)

#drop rows with wrong 'FRANCHISE NAME' 'START' 'END'
df_nf.drop([0, 3, 6, 13, 15, 17, 20, 23, 25, 30, 34, 37, 41, 45, 47], inplace=True)
df_nf.reset_index(drop=True, inplace=True)

df_nf
```
