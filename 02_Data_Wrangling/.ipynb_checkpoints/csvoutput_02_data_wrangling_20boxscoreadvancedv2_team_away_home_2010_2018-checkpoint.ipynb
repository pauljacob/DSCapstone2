```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python
#add game_date

#add team_id_away and team_id_home: df_empty_tiah


df_tbsa1018_tiah = pd.merge(df_game1018[['game_id', 'game_date', 'team_id_away', 'team_id_home']], df_tbsa0618.rename(columns={'GAME_ID':'game_id'}),
         on=['game_id'], how='inner')
p(df_tbsa1018_tiah)
```

```python
#get _away nba advanced boxscores into a dataframe

#filter for away games from 'df_tbsa1018_tiah'
df_away = df_tbsa1018_tiah[df_tbsa1018_tiah.TEAM_ID == df_tbsa1018_tiah.team_id_away].drop(
    columns=['team_id_away', 'team_id_home'])
df_away = df_away.sort_values('game_date').reset_index(drop=True)
p(df_away)
```

```python
#rename columns of away games
```

```python
away_cols = list(df_away.columns)
print(away_cols)
```

```python
away_home_cols = [k for k in away_cols if k not in ['game_id', 'game_date']]
away_home_cols
```

```python
for column in away_home_cols:
    df_away.rename(columns={column:(column+'_away')}, inplace=True)
```

```python
df_away_home_gin = pd.concat([df_away_gin, df_home_gin])
p(df_away_home_gin)
```

```python
df_away_home_gin = df_away_home_gin.groupby('game_id').sum()
df_away_home_gin.reset_index(inplace=True)
p(df_away_home_gin)
```

```python
#concat 'df_away_ginn' and 'df_home_ginn'
pd.concat([df_away_home_gin, df_away_ginn, df_home_ginn], axis=1)
```

```python
#sort each dataframe 'df_away_ginn', 'df_home_ginn', and 'df_away_home_gin' by game_id

df_away_ginn = df_away_ginn.sort_values('game_id').reset_index(drop=True)
p(df_away_ginn)
```

```python
df_home_ginn = df_home_ginn.sort_values('game_id').reset_index(drop=True)
p(df_home_ginn)
```

```python
df_away_home_gin = df_away_home_gin.sort_values('game_id').reset_index(drop=True)
p(df_away_home_gin)
```
