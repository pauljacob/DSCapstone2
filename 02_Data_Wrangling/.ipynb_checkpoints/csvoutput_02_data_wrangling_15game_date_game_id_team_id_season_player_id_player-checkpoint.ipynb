```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sqlite3

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.set_option('max_colwidth', None)
```

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python
os.getcwd()
```

```python
df_bsa = rcp('A12bsa_game_id_team_id_player_id_2009_2018.csv', ['game_date'])
p(df_bsa)
```

```python
df_bsa.columns
```

```python
df_bsa_gdgitispip = df_bsa[['game_date', 'game_id', 'team_id', 'season', 'PLAYER_ID', 'PLAYER_NAME']].rename(columns={'PLAYER_NAME':'player'}).drop_duplicates().reset_index(drop=True)
p(df_bsa_gdgitispip)
```

```python
df_inactive = rcp('A11inactive_game_id_team_id_player_id_2010_2018.csv', ['game_date'])
p(df_inactive)
```

```python
df_inactive.columns
```

```python
df_inactive_gdgitispip = df_inactive[['game_date', 'game_id', 'TEAM_ID', 'season', 'PLAYER_ID', 'player']].rename(columns={'TEAM_ID':'team_id'}).drop_duplicates().reset_index(drop=True)
p(df_inactive_gdgitispip)
```

```python
df_merged = pd.concat([df_bsa_gdgitispip, df_inactive_gdgitispip]).drop_duplicates().reset_index(drop=True)
p(df_merged)
```

```python
df_merged.sort_values('game_date', inplace=True)
```

```python
p(df_merged)
```

```python
df_merged.reset_index(drop=True, inplace=True)
```

```python
p(df_merged)
```

```python
datapath = os.path.join('..', 'processed_data')

if not os.path.exists(datapath):
    os.mkdir(datapath)

datapath_df = os.path.join(datapath, '15game_date_game_id_team_id_season_player_id_player_2010_2018.csv')
if not os.path.exists(datapath_df):
    df_merged.to_csv(datapath_df, index=False)
```
