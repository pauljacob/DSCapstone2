```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

import time
```

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python

#filter for _away games
    # dropping columns ['is_home']
    # - excluding 'game_id' specific columns ['season', 'season_type', 'season_year']
    
    #change their column names to include _away
    

#game_id, wl_away dataframe made for merge before export: df_giwla



```

```python
df_away = df_merged[df_merged.is_home == 'f'].drop(columns=['is_home','season_year', 'season_type', 'season'])
df_away = df_away.sort_values('game_date').reset_index(drop=True)
df_away
```

```python
numeric_col_names = list(df_away.select_dtypes('number').drop(columns=['game_id']).columns)
numeric_col_names
```

```python
numeric_col_names.append('wl')
numeric_col_names_pwl = numeric_col_names
```

```python
numeric_col_names_pwl
```

```python
for column in numeric_col_names_pwl:
    df_away.rename(columns={column:(column+'_away')}, inplace=True)
```

```python
#game_id, wl_away dataframe made for merge before export: df_giwla
df_giwla = df_away[['game_id', 'wl_away']]
p(df_giwla)
```

```python
df_away.drop(columns=['wl_away'], inplace=True)
```

```python
#filter for _home games
    # dropping columns 'is_home'
    # - excluding 'game_id' specific columns 'season', 'season_type', 'season_year'

    #change their column names to include _home
    

#game_id, wl_home dataframe made for merge before export: df_giwlh

```

```python
#filter for _home games
    # dropping columns 'is_home'
    # - excluding 'game_id' specific columns 'season', 'season_type', 'season_year'


df_home = df_merged[df_merged.is_home == 't'].drop(columns=['is_home', 'season_year', 'season_type', 'season'])
df_home = df_home.sort_values('game_date').reset_index(drop=True)
p(df_home)
```

```python
#change their column names to include _home
```

```python
for column in numeric_col_names_pwl:
    df_home.rename(columns={column : (column + '_home')}, inplace=True)
```

```python
df_giwlh = df_home[['game_id', 'wl_home']]
p(df_giwlh)
```

```python
df_home.drop(columns=['wl_home'], inplace=True)
```

```python
#concat df_away and df_home dataframes

#fillna numeric columns with 0 

#groupby game_id and sum()

```

```python
df_away_home = pd.concat([df_away, df_home])
p(df_away_home)
```

```python
df_away_home = df_away_home.fillna(0)
p(df_away_home)
```

```python
df_away_home = df_away_home.groupby('game_id').sum().reset_index()
p(df_away_home)
```

```python
# 'game_id','game_date','season_type','season','season_year' to be saved into a dataframe to be merged
# at the end

#change float64 to int where appropriate
```

```python
# 'game_id','game_date','season_type','season','season_year' to be saved into a dataframe to be merged
# at the end

df_gigdstssy = df_merged[['game_id', 'game_date', 'season_type', 'season', 'season_year']].drop_duplicates()
p(df_gigdstssy)
```

```python
df_away_home = pd.merge(df_away_home, df_gigdstssy, on='game_id', how='inner')
p(df_away_home)
```

```python
df_away_home['w_pct_home'].value_counts().sort_values(ascending=False)
```

```python
#change float64 to int where appropriate
```

```python
columns_away_home = list(df_away_home.columns)
columns_away_home
```

```python
columns_away_home_int = [k for k in columns_away_home
                         if k not in ['w_pct_away', 'fg_pct_away', 'fg3_pct_away','ft_pct_away',
                                      'w_pct_home', 'fg_pct_home', 'fg3_pct_home', 'ft_pct_home',
                                      'game_date', 'season_type', 'season', 'season_year']]
columns_away_home_int
```

```python
df_away_home[columns_away_home_int] = df_away_home[columns_away_home_int].astype('int64')
p(df_away_home)
```

```python
#add df_giwla and df_giwlh back to df_home_away by merge

df_away_home = pd.merge(df_away_home, df_giwla, on=['game_id'], how='inner')
p(df_away_home)
```

```python
df_away_home = pd.merge(df_away_home, df_giwlh, on=['game_id'], how='inner')
p(df_away_home)
```

```python
#calculate away team spread and home team spread

df_away_home['spread_home'] = df_away_home['pts_away'] - df_away_home['pts_home']
```

```python
df_away_home['spread_away'] = df_away_home['pts_home'] - df_away_home['pts_away']
```
