```python
import pandas as pd
```

```python
#game dates and stats
df = pd.DataFrame({'a':[1, 2, 3, 4, 5, 6], 'b':[1, 2, 3, 4, 0, 0]})
df
```

```python
#injury dates
df2 = pd.DataFrame({'a':[1, 2.5, 3, 4, 5, 9], 'b':[1, 2, 3, 4, 5, 6]})
df2
```

```python
pd.merge(df.a.drop_duplicates(), df2, on='a', how='inner')
```
