```python
seasons = ['2009-10', '2010-11', '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18']
seasons
```

```python
first = []
```

```python
seasons[1]
```

```python
last = []
```

```python
for i in range(len(seasons)):
    first.append(df2[df2.season == seasons[i]].game_date.min())
    last.append(df2[df2.season == seasons[i]].game_date.max())
```

```python
first
```

```python
last
```

```python
df_7sfgdlgd = pd.DataFrame({'season': seasons, 'first': first, 'last':last})
df_7sfgdlgd
```

```python
df_7sfgdlgd.info()
```

```python
datapath = os.path.join('..', 'processed_data')

if not os.path.exists(datapath):
    os.mkdir(datapath)

datapath_df = os.path.join(datapath, '7season_first_game_date_last_game_date.csv')
if not os.path.exists(datapath_df):
    df_7sfgdlgd.to_csv(datapath_df, index=False)
```
