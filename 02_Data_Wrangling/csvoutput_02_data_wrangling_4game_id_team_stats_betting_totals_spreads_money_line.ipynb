```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
#return the first 5 and last 5 rows of this dataframe
def p(df_):
    if df_.shape[0] > 6:
        print(df_.shape)
        return pd.concat([df_.head(), df_.tail()])
    else:
        return df_

def rcp(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'processed_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'processed_data', file_), parse_dates=pd_)
    
def rcr(file_, pd_=None):
    if pd_ == None:
        return pd.read_csv(os.path.join('..', 'raw_data', file_))
    else:
        return pd.read_csv(os.path.join('..', 'raw_data', file_), parse_dates=pd_)
```

```python
df_betting_totals = rcr('nba_betting_totals.csv')
p(df_betting_totals)
```

```python
#get at least one row for each game_id
df_betting_totals = df_betting_totals.iloc[df_betting_totals.game_id.drop_duplicates().index].reset_index(drop=True)
p(df_betting_totals)
```

```python
df_game = rcp('3game_date_home_away.csv', ['game_date'])
#filter for seasons 2010-11 thru 2017-18
df_game1018 = df_game[(df_game.season == '2010-11') | 
                      (df_game.season == '2011-12') | 
                      (df_game.season == '2012-13') | 
                      (df_game.season == '2013-14') | 
                      (df_game.season == '2014-15') | 
                      (df_game.season == '2015-16') | 
                      (df_game.season == '2016-17') | 
                      (df_game.season == '2017-18')].reset_index(drop=True)
p(df_game1018)
```

```python
#df_merge = pd.merge(df_game1018, df_betting, on='game_id', how='outer', indicator=True)
df_merge = pd.merge(df_game1018, df_betting_totals, on=['game_id'], how='inner')
p(df_merge)
```
